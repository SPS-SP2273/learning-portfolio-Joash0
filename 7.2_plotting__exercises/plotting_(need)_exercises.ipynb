#### Exercise 1 (Fill in the blanks)

```python
from matplotlib import pyplot as plt

# Some data for plotting
x = [0, 1, 2, 3, 4, 5]
y = [0, 4, 8, 12, 16, 20]
y_err = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5]
x_err = .25

plt.errorbar(x, y, 
             xerr=x_err, yerr=y_err,
             color='brown', linestyle='dashed',
             marker='o', markerfacecolor='yellow',
             ecolor='black',capsize=5,
             label='Y values with errors')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='lower right')
plt.show()
```

#### Exercise 2 (A simple mathematical plot)

```python
from matplotlib import pyplot as plt
import numpy as np
x_values = [x for x in np.linspace(0, 50, 10000)]
y_values = [np.exp(-x/10)*np.sin(x) for x in x_values]
plt.plot(x_values, y_values, color='blue', linestyle='solid')
plt.xlabel('x', fontsize = 15, color = 'darkorange')
plt.ylabel('y', rotation=0, fontsize = 15, color = 'darkorange')
plt.grid(alpha=.25)
plt.title(r'$y=e^{-x/10}\sin(x)$', fontsize = 20)
plt.ylim([-1,1])

```
