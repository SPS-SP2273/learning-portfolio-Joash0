```python
import math
import random
import matplotlib.pyplot as plt

# y = mx + c

def my_line(m, c):
    x_values = [x for x in range(100)]
    y_values = [m*x + c + 50*random.random() for x in x_values]
    return x_values, y_values

x, y = my_line(3, 2)
plt.plot(x, y)
```
