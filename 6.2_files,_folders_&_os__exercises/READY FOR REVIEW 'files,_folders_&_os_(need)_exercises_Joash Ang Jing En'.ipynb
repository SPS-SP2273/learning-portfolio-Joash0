```python
def rename_my_file(old_file_name):
    file_type = old_file_name.split(".")[-1]
    remainder = old_file_name.split(".")[0]
    month_day, city = remainder.split("_")   
    month, day = month_day.split("-")       
    month_num = months[month]
    new_filename = f"{month_num}-{day}_{city}.{file_type}"
    return new_filename

print(rename_my_file(old_file_name))
```

#### Task 7

```python
import glob, os, shutil
import numpy as np

cities = ['Cairo', 'Dhaka', 'Jakarta', 'Karachi', 'Manila',
 'Melbourne', 'Osaka', 'Shanghai', 'Singapore', 'Tokyo']

file_list = []
for city in cities:
    for file_name in glob.glob(f"os-collaboration-exercise-data/**/*_{city}.txt", recursive=True):
        formated_date_city = file_name.split("\\")[-1]
        file_list.append(formated_date_city)
print(file_list)

def finding_duplicate_file(file_name):
    file_np_array = np.array(file_list)
    formated_date_city = file_name.split("\\")[-1]
    mask = file_np_array == formated_date_city
    print(mask.sum())
    return mask.sum()

for city in cities:
    for i in glob.glob(f"os-collaboration-exercise-data/**/*_{city}.txt", recursive=True):
        try:
            folder_name = os.path.join("os-collaboration-exercise-data", city)
            shutil.move(i, folder_name)
        except:
            number = finding_duplicate_file(i)
            if number > 1:
                formated_date_city = i.split("\\")[-1]
                file_list.remove(formated_date_city)
                os.remove(i)
            else:
                pass

# FINALLY LAH I'M DONE :')           
```
